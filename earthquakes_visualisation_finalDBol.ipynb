```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Count and list places with 25  most event occurences 
place_count = equake_data["place"].value_counts()
print(place_count.head(25))
# all US based
```

```python
# Sort date list to investigate if anything notable by number of event occurences

# Count event occurences per date - out of date order
date_count = equake_data["date_extracted"].value_counts()

# Convert to DataFrame 
datetable = pd.DataFrame(date_count)

# Display the result ordered by date
print(datetable.sort_values('date_extracted'))

# reasonably consistent flux of occurences between 200-500 per day except last day in set, likely due to data not logged yet
```

```python
# Sort full dataset by magnitude to look for anything notable
equake_data.sort_values('mag',ascending=False).T

# Global entries at highest magnitude and predominantly American at lower end of table
```

```python
# Sort full dataset by depth to look for anything notable
equake_data.sort_values('depth',ascending=False).T

# Global entries at deepest depth and predominantly American at end of table
```

```python
# Generate new dataset with predominantly only North and South American data using latitudes

#  exclude the longitudes not incorporating north and south America ie. everything to east and west of Americas
# latitude limits (-180, 180)
#   "~" means you want to keep those with the opposite of your condition. The ":" means you want to keep all the columns
# | = "or"
nrth_sth_am_locations = equake_data.loc[~((equake_data['longitude'] <= -165) | (equake_data['longitude'] >= -35)),:]
# Convert to DataFrame 
nrth_sth_am_table = pd.DataFrame(nrth_sth_am_locations).reset_index()

# Display the result
nrth_sth_am_table.T
```

```python
# Count and list places with 25  most event occurences in the Americas
am_place_count = nrth_sth_am_table["place"].value_counts()
print(am_place_count.head(25))
```

```python
# Generate new dataset with predominantly non North and South American or rest of the world data using latitudes

# exclude the longitudes incorporating north and south America ie. the latitudes where the Americas are within
# latitude limits (-180, 180)
#   "~" means you want to keep those with the opposite of your condition. The ":" means you want to keep all the columns
restofworld_locations = equake_data.loc[~((equake_data['longitude'] >= -165) & (equake_data['longitude'] <= -35)),:]
# Convert to DataFrame 
restofworld_table = pd.DataFrame(restofworld_locations).reset_index()

# Display the result
restofworld_table.T
```

```python
# Count and list places with 25  most event occurences outside the Americas, in the rest of the world
rofwld_place_count = restofworld_table["place"].value_counts()
print(rofwld_place_count.head(25))

# still some Alaskan entries due to closeness to Asia and cut-off latitudes, particularly Rat Islands off west coast - most of Alaska should be included in the Americas dataset
```

```python
# For non Americas, rest of world data: Count and list unique values in each column to examine data varability, only if data type in each column is a string
# value counts if datatype is a string
for i in restofworld_table:
    data_type =  restofworld_table[i].dtype
    if data_type == "object":
        value_count = restofworld_table[i].value_counts()
        print("\n",value_count)
# only event type recorded is earthquakes, less networks involved in data recording and review (net, locationSource, magSource) and all data reviewed
```

```python
# Investigate if a pivot table can bring some understanding to string data for full data set
# Create the pivot table using 'count'number of occurences for aggfunc
pivot_table = pd.pivot_table(equake_data, values=["mag"], index=["net","locationSource","magSource","type","magType","status"], aggfunc='count')

# Convert to DataFrame 
net_table = pd.DataFrame(pivot_table)

# Display the result
net_table

# Network information (net, magSource and locationSource) shou net and locationSource correlate with more variation in magSource. 
# Variable magTypes but these are noted as different functions to determine magnitude depending on several conditions 
```

```python
# Investigate if a pivot table can bring some understanding to string data for non Americas dataset, which has less data
# Create the pivot table using 'count'number of occurences for aggfunc and seperate place data by column 
rofwld_pivot_table = pd.pivot_table(restofworld_table, values=["mag"], index=["net","locationSource","magSource","type","magType","status"],columns=["place"], aggfunc='count')

# Convert to DataFrame 
rofwld_net_table = pd.DataFrame(rofwld_pivot_table)

# Display the result
rofwld_net_table

# only earthquake event types and only 3 networks involved (ak, av, us)
```
